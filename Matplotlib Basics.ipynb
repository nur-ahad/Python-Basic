```python
# Normal Workflow in Matplotlibe 

import matplotlib.pyplot as plt
```

```python
# Prepare data 

x = [1,2,3,4,5]
y = [11,22,33,44,55]
```

```python
# Setting up plot -selecting figures and axes 

fig, ax = plt.subplots(figsize=(10,10))
```

```python
# Setting area of the figure 
ax.plot(x,y)

ax.set(title="Sample Data", xlabel="x-axis", ylabel = "y-axis")
fig.savefig("C:/Users/nural/OneDrive/Desktop/Education Documents/simple-plot.png")
```
